```python
# Abrir excel y generar df_coches
df_coches = pd.read_excel('coches.xlsx')
display("DataFrame cargado desde 'coches.xlsx':")
display(df_coches)



# Añade una fila al DF
nueva_fila = {'Marca de coche': 'OPEL', 'Matrícula': 'PDS', 'Velocidad Màxima': 165, 'Origen': 'Holanda'}
df_coches = pd.concat([df_coches, pd.DataFrame([nueva_fila])], ignore_index=True)
display("DataFrame con la nueva fila añadida:")
display(df_coches)

#  código
nueva_coche = pd.DataFrame({'Marca de coche': ['MG'], 'Matrícula': ["DFS"], 'Velocidad Màxima': [145]})
df_coches = pd.concat([df_coches, nueva_coche], ignore_index=True)
display("DataFrame después de concatenar 'nueva_coche':")
display(df_coches)

# Cambiar valor nan por rumania
df_coches['Origen'] = df_coches['Origen'].fillna('Rumania')
display("DataFrame con valores NaN en 'Origen' reemplazados por 'Rumania':")
display(df_coches)

# Exportar el df_coches a un csv
df_coches.to_csv('coches2.csv', index=False)
display("DataFrame exportado a 'coches2.csv':")
```

6. Añade una columna al dataframe llamada "Categoría_Velocidad".
* Esta columna debe categorizar la velocidad máxima de cada coche en 'Baja', 'Media' o 'Alta'.
* Utiliza una función lambda directamente dentro del método `apply()` para lograr esto.
* Define las categorías de velocidad de la siguiente manera:
    * Baja: Velocidad Máxima menor que 170.
    * Media: Velocidad Máxima entre 170 y 190 inclusive.
    * Alta: Velocidad Máxima mayor que 190.

```python

df_coches['Categoría_Velocidad'] = df_coches['Velocidad Màxima'].apply(lambda x: 'Baja' if x < 170 else ('Media' if 170 <= x <= 190 else 'Alta'))

display("DataFrame con la nueva columna 'Categoría_Velocidad':")
display(df_coches)
```

7. Escribe una función que permita a los usuarios filtrar el dataframe por marca y velocidad mínima.

```python
# Escribe una función que permita a los usuarios filtrar el dataframe por marca y velocidad mínima.
def filtrar_por_marca_y_velocidad_minima(df, marca, velocidad_minima):
    """
    Filtra un DataFrame de coches por marca y velocidad mínima.

    Args:
        df (pd.DataFrame): El DataFrame a filtrar.
        marca (str): La marca de coche por la que filtrar.
        velocidad_minima (int): La velocidad mínima para filtrar.

    Returns:
        pd.DataFrame: Un DataFrame filtrado que contiene coches de la marca
                      especificada con una velocidad máxima mayor o igual a la
                      velocidad mínima.
    """
    df_filtrado = df[(df['Marca de coche'] == marca) & (df['Velocidad Màxima'] >= velocidad_minima)]
    return df_filtrado

```
